\section*{Problem 5:}Use the result of Problem 1 and Python to create a plot of this probability, and note after which n the probability becomes more than 50\%.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def no_shared_birth(n):
    p = 1
    for i in range(n):
        p *= (365-i)/365
    return p
n = 50
no_shared_birth(n)
```

```python
1 - no_shared_birth(n)
```

```python
n_tot = list(range(1,101))
probs = [1 - no_shared_birth(n) for n in n_tot] 
```

```python
at_fifty_percent = next(n for n, prob in enumerate(probs, start=1) if prob > 0.5)
at_fifty_percent
```

```python
plt.figure(figsize=(10,6))
plt.plot(n_tot,probs,label="Atleast 2 share same birthday",color='green')
plt.axhline(y=0.5,color='red',linestyle='--')
plt.axvline(x=at_fifty_percent,linestyle='--',color='blue')
plt.title('Atleast two students share the same birthday.')
plt.xlabel('Number of students')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)
plt.show()
```

\section*{Problem 6} Use the result of Problem 3 and Python to create a plot of this probability, and note after which n the probability becomes less than 1\%.

```python
def probablitiy(n):
    return 6*(5/6)**n - 15*(4/6)**n + 20*(3/6)**n - 15*(2/6)**n + 6*(1/6)**n
```

```python
n_values = np.arange(1,1000)
```

```python
probs = []
threshold = 0.01
at_which_n_value = None
find_n_1 = []
for n in n_values:
    P = probablitiy(n)
    probs.append(P)
    find_n_1.append((P,n))
```

```python
value = [n for p,n in find_n_1 if p < 0.01][0]
print("n at which probablity becomes less that 1% is ",value)
```

```python
plt.figure(figsize=(10,4))
plt.plot(n_values,probs)
plt.axhline(y = threshold, color='red',linestyle='--')
plt.axvline(x=value,color='green',linestyle='--')
plt.title('Probabilites vs n dice rolls')
plt.xlabel("n")
plt.ylabel('Probabilities')
plt.show()
```
